```python
class LinearShiftedColormap(Colormap):
    '''
    Returns a new colormap based on the given one such that
    newcmap = oldcmap(shift + (1-shift) * a)
    hardzero maintains newcmap(0) = oldcmap(0) and then discontinuously jumps. Hardzero doesn't work on iterables because I'm lazy.
    '''
    def __init__(self, cmap, shift, hardzero=True):
        self.cmap = cmap
        self.shift = shift
        self.hardzero = hardzero
        
    def __call__(self, xi, **kw):
        if self.hardzero and not hasattr(xi, '__iter__') and xi == 0.0:
            yi = 0
        else:
            yi = self.shift + (1 - self.shift) * xi
        return self.cmap(yi, **kw)

    def __getattr__(self, attr):
        return getattr(self.cmap, attr)
```
